```python
from sympy import *
from sympy import symbols
import holoviews as hv 
from holoviews import streams
from holoviews import opts
import numpy as np
```

```python
r,a,b = symbols('r a b')

lennardJones = a/r**12 - b/r**6; 
lennardJones
force = diff(lennardJones,r)
force
```

```python
rMin , eMin = symbols('rMin eMin')
equationDistanceMinimum = force.subs(r,rMin)
print(equationDistanceMinimum)
equationEnergieMinimum = lennardJones.subs(r,rMin) + eMin
print(equationEnergieMinimum)
criticalPointSolutions = solve([equationEnergieMinimum,equationDistanceMinimum])
print(criticalPointSolutions)
```

```python
def substitute(var,critical_sol,equ):
    return equ.subs(var,critical_sol[0][var])

lennardJonesCritical = substitute(b,criticalPointSolutions,lennardJones)
lennardJonesCritical = substitute(a,criticalPointSolutions,lennardJonesCritical)
print(lennardJonesCritical)
lennardJonesScaled = expand(lennardJonesCritical/eMin)
print(lennardJonesScaled)
```

```python
p,p_ = symbols('p p_')
energy = {}
energy[p_] = lambda x: lennardJonesScaled.subs(r,x*rMin)
energy[p_](p_)
forceScaled = {}
forceScaled[p_] = lambda x: diff(energy[p_](p),p).subs(p,x)
forceScaled[p_](p_)
forceScaledAttractive = expand(forceScaled[p_](p_)).as_terms()[0][1][0]
forceScaledRepulsive = expand(forceScaled[p_](p_)).as_terms()[0][0][0]
print(forceScaledAttractive)
print(forceScaledRepulsive)
energyAttractive = expand(energy[p_](p_)).as_terms()[0][1][0]
print(energyAttractive)
energyRepulsive = expand(energy[p_](p_)).as_terms()[0][0][0]
print(energyRepulsive)
```

```python
force_np = lambdify(p_,forceScaled[p_](p_),"numpy")
force_attractive_np = lambdify(p_,forceScaledAttractive,"numpy")
force_repulsive_np = lambdify(p_,forceScaledRepulsive,"numpy")

energy_attractive_np = lambdify(p_,energyAttractive,"numpy")
energy_repulsive_np = lambdify(p_,energyRepulsive,"numpy")
energy_np = lambdify(p_,energy[p_](p_),"numpy")
```

```python
xs = np.linspace(0.8, 2, 1000)
size_point = 10
e_min = (1,-1)
f_min = (1,0)

def function_force(t):
    return force_np(t)

def function_energy(t):
    return energy_np(t)

def force_point(dist):
    coords = (dist,force_np(np.array(dist)))
    point = hv.Points(coords).opts(size=size_point,color='gray',alpha=0.8)
    return point*hv.Text(coords[0],coords[1]+size_point/10, 'p=%.2f,E=%.2f' % (coords[0] ,coords[1]))

def energy_point(dist):
    coords = (dist,function_energy(np.array(dist)))
    point = hv.Points(coords).opts(size=size_point,color='gray',alpha=0.8)
    return point*hv.Text(coords[0],coords[1]+size_point/10, 'p=%.2f,F=%.2f' % (coords[0] ,coords[1]))

xDim = hv.Dimension('Distance Interatomique normalisee')
yDimEnergy = hv.Dimension('Energy Potentiellenormalisee',range = (-2,10))
yDimForce = hv.Dimension('Force, F ',range =(-30,10))
curve_force = hv.Curve((xs,function_force(xs)),label = 'total force').redim(y=yDimForce).redim(x =xDim).opts(title='force',color='red')
curve_force_attr = hv.Curve((xs,force_attractive_np(xs)),label = 'attractive').redim(y=yDimForce).redim(x =xDim).opts(title='force',color='blue')
curve_force_rep = hv.Curve((xs,force_repulsive_np(xs)),label = 'repulsive').redim(y=yDimForce).redim(x =xDim).opts(title='force',color='blue')

min_force_pt = hv.Points(f_min).opts(size=size_point,color='red',alpha=0.8).redim(y=yDimForce).redim(x =xDim)#*hv.Arrow(f_min[0], f_min[1], 'F min', '^').redim(y=yDimForce).redim(x =xDim)
min_energy_pt = hv.Points(e_min).opts(size=size_point,color='red',alpha=0.8)#*hv.Arrow(e_min[0], e_min[1], 'E min', 'v')
curve_energy = hv.Curve((xs,function_energy(xs)),label='total energy').redim(y=yDimEnergy).redim(x =xDim).opts(title='energy',color='red',alpha=0.8)
curve_energy_attr = hv.Curve((xs,energy_attractive_np(xs)),label='attractive').redim(y=yDimEnergy).redim(x =xDim).opts(title='energy',color='blue',alpha=0.8)
curve_energy_rep = hv.Curve((xs,energy_repulsive_np(xs)),label='repulsive').redim(y=yDimEnergy).redim(x =xDim).opts(title='energy',color='green',alpha=0.8)

crosshair = hv.DynamicMap(force_point, kdims='dist').redim.range(dist=(.8,2.)).redim.default(dist=1.5).opts(width=300,height=600)
crosshair2 = hv.DynamicMap(energy_point, kdims='dist').redim.range(dist=(.8,2.)).redim.default(dist=1.5).opts(width=300,height = 600)
(curve_force * crosshair*curve_energy_rep*curve_force_attr*min_force_pt).opts(legend_position='bottom') + (curve_energy*curve_energy_attr*crosshair2*curve_energy_rep).opts(legend_position='bottom')
```

```python
matrial_energy = lambda epsilon, r0, r: epsilon * ((r0/r)**12 - 2 * (r0/r)**6)
size_point = 10
e_min = (2,-1)
f_min = (1,0)

def material(eps0_1, r0_1, eps0_2, r0_2):
    xs = np.linspace(0.8, 2, 1000)
    xDim = hv.Dimension('Distance Interatomique r')
    yDimEnergy = hv.Dimension('Energy Potentiellenormalisee E [kcal/mol]',range = (-3,3))
    curve_material_1 = hv.Curve((xs, matrial_energy(r=xs, epsilon=eps0_1, r0=r0_1)),label = 'material 1').redim(y=yDimEnergy).redim(x =xDim).opts(color='red')
    curve_material_2 = hv.Curve((xs, matrial_energy(r=xs, epsilon=eps0_2, r0=r0_2)),label = 'material 2').redim(y=yDimEnergy).redim(x =xDim).opts(color='blue')
    return curve_material_1 * curve_material_2

curve_material = hv.DynamicMap(material, kdims=['eps0_1', 'r0_1', 'eps0_2', 'r0_2']).redim.range(eps0_1=(.1, 2.), 
                                                                                                 r0_1=(.1, 2.), eps0_2=(.1, 2.), 
                                                                                                 r0_2=(.1, 2.)).redim.default(eps0_1=1., r0_1=1., eps0_2=1., 
                                                                                                                              r0_2=1.)

curve_material

```

```python
approxLennardJonesWithUnits = lambda eps0, r0, r: -eps0 + (36 * (r - r0)**2 * eps0) / r0**2
approxForceWithUnits = lambda eps0, r0, r: (72 * (r - r0) * eps0) / r0**2
lennardJonesWithUnits = lambda eps0, r0, r: (-((2* r0**6)/r**6) + r0**12/r**12) * eps0
forceWithUnits = lambda eps0, r0, r: ((12 * r0**6)/r**7 - (12 * r0**12)/r**13) * eps0


def force_potential(eps0, r0):
    xs = np.linspace(0.5, 2, 1000)
    xDim = hv.Dimension('Distance Interatomique r')
    yDimEnergy = hv.Dimension('Energy Potentiellenormalisee E [KJ/mol]& Force F[N]',range = (-3,5))
    curve_approxLennardJonesWithUnits = hv.Curve((xs, approxLennardJonesWithUnits(r=xs, eps0=eps0, r0=r0)),label = 'Approximation parabolique du potentiel LJ').redim(y=yDimEnergy).redim(x =xDim).opts(color='red')
    curve_approxForceWithUnits = hv.Curve((xs, approxForceWithUnits(r=xs, eps0=eps0, r0=r0)),label = 'Force du potential approximé').redim(y=yDimEnergy).redim(x =xDim).opts(color='blue')
    curve_lennardJonesWithUnits = hv.Curve((xs, lennardJonesWithUnits(r=xs, eps0=eps0, r0=r0)),label = 'Potentiel de LJ').redim(y=yDimEnergy).redim(x =xDim).opts(color='green')
    curve_forceWithUnits = hv.Curve((xs, forceWithUnits(r=xs, eps0=eps0, r0=r0)),label = "Force LJ").redim(y=yDimEnergy).redim(x =xDim).opts(color='yellow')
    equ = hv.Text(1.75, 1, 'E = 72*($ε_0^2$)/($r_0^2$) = %.2f'%(72*eps0**2/r0**2))
    return curve_approxLennardJonesWithUnits * curve_approxForceWithUnits * curve_lennardJonesWithUnits * curve_forceWithUnits * equ

material_force_potential = hv.DynamicMap(force_potential, kdims=['eps0', 'r0']).redim.range(eps0=(.2, 5.), r0=(.7, 2.)).opts(legend_position='right')

                                                                                                
material_force_potential
```

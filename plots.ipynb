```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
directory = 'data/'
questionnairs = dict()

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith('csv'):
        questionnairs[filename] = {}
        with open(directory + filename) as f:
            lines = f.readlines()
            for line in lines:
                ans = "".join(line.split(',')[1:]).strip("\n")
                try:
                    questionnairs[filename][line.split(',')[0]] = int(ans)
                except:
                    questionnairs[filename][line.split(',')[0]] = ans
                
```

```python
isinstance(questionnairs['response_wanhao_1.csv']['1.1'], int)
```

```python
anws = dict()
for _, questionnair in questionnairs.items():
    for question in questionnair.keys():
        if question not in anws.keys():
            anws[question] = []
        anw = questionnair[question]
        if isinstance(anw, int):
            anws[question].append(anw)
```

```python
for _, question in anws.items():
    if question:
        
        figure = plt.figure(figsize=(8, 8))
        ax = figure.add_subplot(111)
        ax.hist(question)
    
```
